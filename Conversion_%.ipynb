```python
import pandas as pd
import numpy as np

def calculate_na_percentage(file_path):
    # Load the dataset
    df = pd.read_csv(file_path)
    
    # 1. Filter for visits with a non-zero number of frames
    # This ensures we only count rows where 'ValidFrames' is greater than 0
    df_valid = df[df['ValidFrames'] > 5].copy()
    
    total_valid_visits = len(df_valid)
    
    if total_valid_visits == 0:
        return "No visits found with more than 0 frames."

    # 2. Identify "N/A" labels
    # In pandas, empty or "N/A" values in a CSV are typically loaded as NaN (Not a Number)
    # We check for both actual NaNs and the literal string "N/A" just in case
    na_condition = df_valid['Predicted_SowID'].isna() | (df_valid['Predicted_SowID'] == 'N/A')
    na_count = na_condition.sum()
    
    # 3. Calculate percentage
    percentage = (1-(na_count / total_valid_visits)) * 100
    
    # Output the exact numbers
    print(f"File: {file_path}")
    print(f"Total visits (ValidFrames > 0): {total_valid_visits}")
    print(f"Visits with non N/A labels: {na_count}")
    print(f"Percentage of N/A visits: {percentage:.2f}%")
    
    return percentage

# Example usage:
# Replace 'your_file_name.csv' with the actual name of your file
file_name = r"C:\Users\spaudel\OneDrive - University of Arkansas System Division of Agriculture\Desktop\Projects_shivapdl\ID_Organized_CVPR\Output_Logs\Day2_test_1\visit_level_logs_Day2_1_10sows.csv"
calculate_na_percentage(file_name)
```
